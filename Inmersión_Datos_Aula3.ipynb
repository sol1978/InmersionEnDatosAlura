```python
import pandas as pd
inmuebles = pd.read_csv ('/content/drive/MyDrive/Inmersion_de_Datos_Latam/inmuebles_bogota.csv')
```

```python
columnas = {'Baños':'Banos', 'Área': 'Area'}
inmuebles = inmuebles.rename (columns = columnas)
inmuebles.describe ()
```

```python
inmuebles ['Valor'] = inmuebles ['Valor'].str.replace ('$','')
inmuebles.head (10)
```

```python
inmuebles ['Valor'] = inmuebles ['Valor'].str.replace ('.','')
inmuebles.head ()
```

```python
inmuebles ['Valor'] =pd.to_numeric (inmuebles['Valor'])
inmuebles
```

```python
inmuebles.info ()
```

```python
inmuebles['Valor_Millon'] =inmuebles.Valor.astype ('float')/1000000
inmuebles.info ()
```

```python
inmuebles.describe ()
```

```python
pd.set_option('display.precision',2)
pd.set_option('display.float_format', lambda x:'%.2f' %x)
inmuebles.describe ()
```

```python
inmuebles ['Valor_m2_Millon'] = inmuebles ['Valor_Millon']/inmuebles['Area']
inmuebles.head (3)
```

```python
inmuebles.groupby ('Barrio').mean ()
```

```python
datos_barrio = inmuebles.groupby ('Barrio').sum ()
datos_barrio
```

```python
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Valor_Millon']/datos_barrio['Area']
datos_barrio
```

```python
m2_barrio = dict (datos_barrio['Valor_m2_Barrio'])
```

```python
inmuebles ['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles ['Valor_m2_Barrio'] = inmuebles ['Valor_m2_Barrio'].map (m2_barrio)
inmuebles.head (10)
```

```python
top_barrios = inmuebles ['Barrio'].value_counts () [:10].index
```

```python
datos_barrio.reset_index (inplace=True)
datos_barrio
```

```python
datos_barrio.query('Barrio in @top_barrios')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure (figsize=(10,8))
ax = sns.barplot (x='Barrio', y='Valor_m2_Barrio', data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params (axis = 'x', rotation =45)
```

```python
plt.figure (figsize=(10,8))
ax = sns.boxplot (x='Barrio', y='Valor_Millon', data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params (axis = 'x', rotation =45)
plt.show ()
```

```python
plt.figure (figsize=(10,8))
ax = sns.boxplot (x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios'))
ax.tick_params (axis = 'x', rotation =45)
plt.show ()
```

```python
plt.figure (figsize=(10,8))
ax = sns.boxplot (x='Barrio', y='Valor_m2_Millon', data = inmuebles.query('Barrio in @top_barrios & Valor_m2_Millon <15'))
ax.tick_params (axis = 'x', rotation =45)
plt.show ()
```

```python
plt.figure (figsize=(10,8))
ax = sns.boxplot (x='Barrio', y='Area', data = inmuebles.query('Barrio in @top_barrios & Area <500'))
ax.tick_params (axis = 'x', rotation =45)
plt.show ()
```

```python
plt.figure (figsize=(10,8))
ax = sns.boxplot (x='Barrio', y='Valor_Millon', data = inmuebles.query('Barrio in @top_barrios & Valor_Millon <2000'))
ax.tick_params (axis = 'x', rotation =45)
plt.show ()
```
