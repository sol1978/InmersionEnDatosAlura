```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure (figsize=(10,6))
grafica = sns.histplot (data=inmuebles, x='Valor_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribución de Valores de los inmuebles en Bogotá')
grafica.set_xlabel ('Valor en millones')
grafica.set_ylabel ('Cantidad de inmuebles')
plt.xlim ((50,1000))
plt.show ()
```

```python
inmuebles_principales = inmuebles[inmuebles['Tipo'].isin(['Casa','Apartamento', 'Oficina/Consultorio'])]
plt.figure(figsize=(10,8))
grafica=sns.histplot(data=inmuebles_principales,x='Valor_Millon',kde=True,hue='Tipo')
grafica.set_title('Distribución de Valores de los inmuebles en Bogotá')
grafica.set_xlabel ('Valor en millones')
grafica.set_ylabel ('Cantidad de inmuebles')
plt.xlim ((50,1000))
plt.savefig ('/content/drive/MyDrive/Inmersion_de_Datos_Latam/valor_inmuebles.png', format = 'png')
plt.show ()
```

```python
inmuebles['Precio_m2'] = inmuebles['Valor_Millon'] / inmuebles['Area']
inmuebles.head(10)
```

```python
precio_m2_inmuebles=inmuebles[['Barrio','Precio_m2']].groupby(['Barrio']).mean().sort_values(by=['Precio_m2'], ascending=False).head(10)

precio_m2_inmuebles.plot(kind='bar')

plt.xlabel('Barrio')
plt.ylabel('Precio promedido del metro cuadrado')
plt.title('Top 10 barrios de mayor precio por metro cuadrado en Colombia')
plt.savefig ('/content/drive/MyDrive/Inmersion_de_Datos_Latam/valor_inmuebles_1.png', format='png')
plt.show()
```

```python
precio_m2_inmuebles=inmuebles[['Barrio','Precio_m2']].groupby(['Barrio']).mean().sort_values(by=['Precio_m2'], ascending=True).head(10)

precio_m2_inmuebles.plot(kind='bar')

plt.xlabel('Barrio')
plt.ylabel('Precio promedido del metro cuadrado')
plt.title('Top 10 barrios de menor precio por metro cuadrado en Colombia')
plt.show()
```
